```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df['Year'], df['Growth Rate'], linestyle='-')
plt.title('Kathmandu Population Growth Rate')
plt.xlabel('Year')
plt.ylabel('Growth Rate %')
plt.grid(True)

plt.show()
```

```python
# bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Growth Rate'], color='skyblue', edgecolor='black')
plt.title('Population Growth Rate by Year')
plt.xlabel('Year')
plt.ylabel('Growth Rate')

plt.xticks(df['Year'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
colors = plt.cm.viridis((df['Growth Rate'] - df['Growth Rate'].min()) /
                        (df['Growth Rate'].max() - df['Growth Rate'].min()))
```

```python
plt.figure(figsize=(10, 4))
bars = plt.bar(df['Year'], df['Growth Rate'], color=colors, edgecolor='black')
```
